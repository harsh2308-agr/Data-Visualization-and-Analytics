### 2. Remove the name and ethnicity column (to ensure confidentiality)

```python
del df1["Name"]
del df1["Ethinicity"]
print("DS Score dataframe")
df1
```

### 3. Fill missing score data with zero

```python
df1["Score"].fillna(0, inplace=True)
display(df1)
```

### 4. Merge two files

```python
merged_df = df1.merge(df1, on="ID", suffixes=(
    '_math', '_ds')).merge(df3, on="ID", suffixes=('_ds', '_physics'))
display(merged_df)
```

```python
result = pd.concat([df1, df2], axis=1, join='inner')
display(result)
```

```python
result1 = pd.concat([result, df3], axis=1, join='inner')
display(result1)
```

### 5. Change Sex(M/F) Column to 1/2 for further analysis 

```python
merged_df["Sex"] = [1 if sex is "M" else 2 for sex in merged_df["Sex"]]
display(merged_df)
```

### 7. Convert ethnicity to numerical value 

```python
df4=pd.read_csv("B_43_Harsh Agrawal - DSScoreTerm1.csv")
Ethinicity_dict=dict()
i=1;
for j in df4["Ethinicity"]:
    if Ethinicity_dict.get(j)==None:
        Ethinicity_dict[j]=i
        i=i+1

print(Ethinicity_dict)    
```

```python
df4["Ethinicity"]=[Ethinicity_dict.get(k) for k in df4["Ethinicity"]]
display(df4)
```

### 8. Fill the missing score for a student to the average of the class

```python
df4["Score"].fillna(df4["Score"].mean(), inplace=True)
display(df4)
```
