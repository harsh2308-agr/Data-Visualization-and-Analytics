### 1. Read Total profit of all months and show it using a line plot

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("B_43_Harsh Agrawal - company_sales_data.csv")
df
```

Inference- Python libraries pandas and matplotlib have been imported and company_sales_data csv file has been read and displayed

```python
x = df['month_number'].tolist()
y = df['total_profit'].tolist()

plt.plot(x,y, marker='d')
plt.title("Total profit of all months")
plt.xlabel("Months")
plt.ylabel("Total Profit")
plt.xticks(x)
plt.show()
```

Inference- The above plotted line chart demonstartes a line chart between total profit vs month number. <br>
A line graph is useful for displaying data or information that changes continuously over time. The line shows a change in data over time.<br> 
Second month has the lowest total profit and eleventh month has the heighet total profit. <br>
A drastic spike has been observed on sixth month.

### 2. Create data using numpy array from 0 to 5 with a gap of 0.01 for x axis, consider y axis as square values of x and show the line graph with grid lines


```python
import numpy as np
x = np.arange(0,5,0.01) 
y = [i*i for i in x]

plt.plot(x,y)
plt.grid(axis='x')
plt.grid(axis='y')
plt.title("y axis as square values of x")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

 A numpy array has been created for values ranging from 0 to 5 with a gap of 0.01 and y axis shows the square of x values.

### 3. Create linear, square and cube data points and create a multiline graph for the same with title to graph and both axis.

```python
import numpy as np
x = np.arange(0,5,1)
y = [i*i for i in x]
z = [i*i*i for i in x]

plt.plot(x,x,x,y,x,z)
plt.title("linear, square and cube data points")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["Linear","Square","Cube"])
plt.xticks(x)
plt.show()
```

A multiline graph has been created with three lines with linear, square and cube data values.
